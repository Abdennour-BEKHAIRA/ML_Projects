```python
#import the needed libraries
import numpy as np ;
import matplotlib.pyplot as py ;
import pandas as pd ;
```

```python
#import the dataset
df = pd.read_csv("/content/Churn_Predictions.csv");
```

```python
df.head
```

```python
x=df.iloc[:,3:13].values;
y=df.iloc[:,13].values;
```

```python
#label encoding for the categoricaal data in our data_set
from sklearn.preprocessing import LabelEncoder,OneHotEncoder;
a=LabelEncoder()
x[:,2]=a.fit_transform(x[:,2]);
```

```python
#hot encoding for the categoricaal data in our data_set
from math import remainder
from sklearn.compose import ColumnTransformer
ct= ColumnTransformer([('ohe',OneHotEncoder(),[1])],remainder='passthrough')
x=np.array(ct.fit_transform(x),dtype=np.str)
x=x[:,1:]
```

```python
#split our DS into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0);
```

```python
#Feature_Scaling 
from sklearn.preprocessing import StandardScaler;
sc = StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#import DL libraries
from tensorflow.keras.models import Sequential
from keras.layers import Dense
#Initialize the ANN 
classifier=Sequential()
#add layers
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu',input_dim=11))
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
#add the output layer
classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
#let's compile our ANN
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#Let's train our model
classifier.fit(x_train,y_train,batch_size=10,epochs=100)
```

```python
#Let's test our model 
y_pred = classifier.predict(x_test)
y_pred=(y_pred>0.5)
```

```python
#see our confusion matrix
from sklearn.metrics import confusion_matrix , accuracy_score
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
#see the accuracy of our model
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
